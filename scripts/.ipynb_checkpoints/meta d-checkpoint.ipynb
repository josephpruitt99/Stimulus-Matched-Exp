```python
datafile = pd.read_csv(data_path/'all_pid_exp1A_clean.csv')
#datafile = pd.read_csv(data_path/'exp1A_1_cleaned.csv')
df = datafile.loc[datafile.check_eye==1]
df = datafile.loc[datafile.check_eye==1]
df = df.rename(columns={'signalOpac': 's', 'answerResp.keys': 'rp', 'slider.response':'rt'})
```

```python
map_resp = {'0':'noise','1':'signal'}
df['response'] = df['rp'].map(map_resp)
df['response'] = df['response'].astype('category')
df['rt'] = df['rt'].astype('category')
df['Stimulus'] = df['Stimulus'].astype('category')
df['position'] = df['position'].astype('category')
df['degrees'] = df['degrees'].astype('category')
```

```python
conf_counts = df.pivot_table(index='participant',columns=['degrees','Stimulus','response','rt'],values='answerResp.corr',aggfunc = 'count')
```

```python
conf_counts=conf_counts+0.25 

# Hautus, M. J. (1995). Corrections for extreme proportions and their biasing 
# effects on estimated values of d'. Behavior Research Methods, Instruments, 
# Computers, 27, 46-51.
```

```python
fit_data = {}
#pos = df.degrees.unique()
pos = df.position.unique()

for p in pos:
        pos_df = conf_counts[p]
        s1 = pos_df['noise'].values[0]
        s2 = pos_df['signal'].values[0]
        
        s1_1 = s1[:4]
        s1_2 = s1[4:]
        
        s1_1 = s1_1[::-1] #reverse the order of rating
        s1 = list(s1_1)+list(s1_2)
        
        s2_1 = s2[:4]
        s2_2 = s2[4:]
        
        s2_1 = s2_1[::-1]
        s2 = list(s2_1)+list(s2_2)
        
        print('--------------------------------------------------')

        print(pos_df.values)
        print(s1,s2)
        fit_data[p] = fit_meta_d_MLE(s1,s2)
```

```python
measures = ['da','meta_da','M_diff','M_ratio','meta_ca','t2ca_rS1','t2ca_rS2','logL']
measure_val = {}
for m in measures:
    measure_val[m] = []
pos_index = []
```

```python
for pos,fit in fit_data.items():
    
    pos_index.append(pos)
    for m in measures:
        measure_val[m].append(fit[m])
```

```python
summary_df = pd.DataFrame(measure_val,index=pos_index)
```
